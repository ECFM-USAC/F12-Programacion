```python
import pandas as pd
import matplotlib.pyplot as plt

# Lectura del archivo local 🤓
ruta = r'C:\Users\ruben\OneDrive\Escritorio\tarea-1\Pokemon.csv'
df = pd.read_csv(ruta)

# Resumen dinámico: muestra tamaño y nombres de columnas 
print(f"Dataset cargado con éxito.")
print(f"Total de registros: {df.shape[0]} | Total de columnas: {df.shape[1]}")

# Verificamos los primeros 10 para ver cómo se llenaron los vacíos 🤓
print("Tabla con los nulos ya procesados:")
display(df[['Name', 'Type1', 'Type2']].head(10))
# 5. Verificación final 🤓
print("\n--- Conteo de nulos tras la limpieza ---")
print(df.isnull().sum())

# Mostramos los primeros registros para ver el cambio 🤓
display(df.head())

```

```python
# Ordenamos los top 10 de hp porque bueno, quiero un pokemon que aguante en mi team 🤓
df_hp = df.sort_values(by='HP', ascending=False)

print("Top 10 Pokémon con mayor cantidad de HP (Puntos de Salud):")
df_hp.head(10)
```

```python

# un buen equipo necesita 1 a 2 atacantes fisicos para poder barrer equipos mirare los top 10 🤓 (en esta casa odiamos las estrategias walleras)
df_ataque = df.sort_values(by='Attack', ascending=False)

# Mostramos los primeros 100 con mayor ataque
print("Top 10 Pokémon con mayor Ataque Físico:")
df_ataque.head(10)
```

```python
#top 10 en defensa fisica, aunque no me gusto las estrategias wall, un buen equipo siemopre necesita al menos 1 o 2 
df_defensa = df.sort_values(by='Defense', ascending=False)

print("Top 10 Pokémon con mayor Defensa Física:")
df_defensa.head(10)
```

```python
#  Sp. Atk top 10, obviamente si me ponen una wall fisica, llevar solo full team fisico sseria bobo
df_sp_atk = df.sort_values(by='Sp. Atk', ascending=False)

print("Top 10 Pokémon con mayor Ataque Especial (Sp. Atk):")
df_sp_atk.head(10)
```

```python
#top 10 def. special.  vere si agrego un defensor especial porque si no me harian pupita
df_sp_def = df.sort_values(by='Sp. Def', ascending=False)

print("Top 10 Pokémon con mayor Defensa Especial (Sp. Def):")
df_sp_def.head(10)
```

```python
# top 10 en velo. para mi estrategia favorita (equipos hiper ofense)🤓 necesito mucha velo, chequemos los mejores para ver de hacer pupita al contrincante 
df_velocidad = df.sort_values(by='Speed', ascending=False)

print("Top 10 Pokémon más rápidos (Speed):")
df_velocidad.head(10)
```

```python
# me causo curiosidad cual generacion es la que tiene mayor estadisticas de "attaque" creo que va ser 6.ta por sus megas 🤓
comparativa_gen = df.groupby('Generation')['Attack'].agg(['mean', 'max', 'std'])

print("Comparativa de Ataque por Generación (Promedio, Máximo y Desviación Estándar):")
display(comparativa_gen)

# Graficamos el promedio y el máximo🤓
comparativa_gen[['mean', 'max']].plot(kind='bar', figsize=(12, 6), color=['skyblue', 'salmon'])

plt.title('Comparativa de Ataque: Promedio vs Máximo por Generación')
plt.xlabel('Generación')
plt.ylabel('Puntos de Ataque')
plt.legend(['Ataque Promedio (mean)', 'Ataque Máximo (max)'])
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.show()
```
